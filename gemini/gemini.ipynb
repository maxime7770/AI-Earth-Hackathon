```python
import google.generativeai as genai
```

```python
import pandas as pd

df = pd.read_csv('AI EarthHack Dataset.csv', encoding='latin-1')
```

```python
df.head()
```

```python
for col in df.columns:
  print(col)
```

```python
problems = df.loc[:,"problem"]
solutions = df.loc[:,"solution"]


# problems_temp = []
# solutions_temp = []
# for i in range(10):
#   problems_temp.append(problems[i])
#   solutions_temp.append(solutions[i])

# print(problems_temp)
# print(solutions_temp)
```

```python
genai.configure(api_key="INSERT_YOUR_KEY")
model = genai.GenerativeModel('gemini-pro')
```

```python
feasability_rankings = []
impact_rankings = []
novelty_rankings = []

for i in range(len(problems)):

  print(i)

  feasability_rankings.append(model.generate_content(f'Given the following input problem-solution pair, return an integer on a scale of 1-3 (inclusive) representing the feasibility of the solution given the problem.\n\nProblem: {problems[i]}\n\nSolution: {solutions[i]}'))
  impact_rankings.append(model.generate_content(f'Given the following input problem-solution pair, return an integer on a scale of 1-3 (inclusive) representing the impact of the solution given the problem.\n\nProblem: {problems[i]}\n\nSolution: {solutions[i]}'))
  novelty_rankings.append(model.generate_content(f'Given the following input problem-solution pair, return an integer on a scale of 1-3 (inclusive) representing the novelty of the solution given the problem.\n\nProblem: {problems[i]}\n\nSolution: {solutions[i]}'))
```

```python
for i in range(len(feasability_rankings)):
  print(f'Feasability: {feasability_rankings[i].text}, Impact: {impact_rankings[i].text}, Novelty: {novelty_rankings[i].text}')
```

```python
#Salvage Code

import numpy as np

feasability_arr = np.array(feasability_rankings)
impact_arr = np.array(impact_rankings)
novelty_arr = np.array(novelty_rankings)
```

```python
#Converting GenAI objects to text

salvaged_feasability_arr = []
salvaged_impact_arr = []
salvaged_novelty_arr = []

for ranking in feasability_rankings:
    salvaged_feasability_arr.append(ranking.text)

for ranking in impact_rankings:
    salvaged_impact_arr.append(ranking.text)

for ranking in novelty_rankings:
    salvaged_novelty_arr.append(ranking.text)

salvaged_feasability_arr.pop()

print(salvaged_feasability_arr)
print(salvaged_impact_arr)
print(salvaged_novelty_arr)
```

```python
output_pd = pd.DataFrame(data=[salvaged_feasability_arr,salvaged_impact_arr, salvaged_novelty_arr]).T
output_pd.columns=['FEASABILITY','IMPACT', 'NOVELTY']
print(output_pd)
```

```python
output_pd.to_csv('salvaged_data.csv')
```
